```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# pie chart

```python
areas=["Marketing","Sales","Devlopment","HR","Customer_support"]
budget=[3.5,4,15,9,2]
```

```python
plt.pie(budget, labels=areas)
```

```python
plt.pie(budget, labels=areas)
plt.show()
```

```python
plt.pie(budget, labels=areas ,autopct="%0.1f%%")
plt.legend()
plt.show()
```

# plot

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[200,400,300,1200,990,876,800,1000]
ofline_price=[800,100,900,500,690,376,200,1200]
chorbazar_price=[1000,200,300,100,890,50,600,1400]

plt.plot(product_id,online_price,color="red", marker="D" ,markersize=10,label="online price")
plt.plot(product_id,ofline_price,color="yellow", marker="+" ,markersize=10,label="ofline price")
plt.plot(product_id,chorbazar_price,color="green",markersize=10,label="chori wale price")
plt.xlabel('product_id')
plt.ylabel('online_price')

plt.title('product price comperison')
plt.legend()
plt.show()
```

# bar chart

```python
year=[2011,2012,2013,2014,2015,2016,2017,2018]
price=[200000,300000,400000,500000,600000,700000,800000,900000]
plt.bar(year,price ,color=["pink","green"])
plt.xlabel("year")
plt.ylabel("price")
plt.title("year by year price comp")
plt.show()
```

# scatter plot

```python
x=[23,45,68,11,25,34,67,98]
y=[26,89,200,15,28,34,72,100]
plt.scatter(x,y ,color="green")
plt.xlabel('age')
plt.ylabel('salary')
plt.title('age vs salary graph')
plt.show()
```
